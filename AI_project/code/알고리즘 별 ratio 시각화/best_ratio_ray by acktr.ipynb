```python
%matplotlib notebook
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# 주어진 각도로 초기 보기 설정
ax.view_init(elev=26, azim=-139)
# 데이터의 최소 및 최대 delivery_id 값을 찾아 색상 매핑에 사용
delivery_ids = [box[7] for box in data[73]]
min_id, max_id = min(delivery_ids), max(delivery_ids)
# 배경 설정
ax.set_xlim(min([box[3] for box in data[73]]), max([box[3] + box[0] for box in data[73]]))
ax.set_ylim(min([box[4] for box in data[73]]), max([box[4] + box[1] for box in data[73]]))
ax.set_zlim(0, max([box[5] + box[2] for box in data[73]]))
# 색상 막대 한 번만 추가
sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=0, vmax=10))
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, label='Delivery ID')
def update(frame):
    ax.cla()  # 이전 프레임의 그래프를 지웁니다.
    ax.view_init(elev=26, azim=-139)  # 각 프레임에서도 동일한 보기 각도 유지
    ax.set_xlim(min([box[3] for box in data[73]]), max([box[3] + box[0] for box in data[73]]))
    ax.set_ylim(min([box[4] for box in data[73]]), max([box[4] + box[1] for box in data[73]]))
    ax.set_zlim(0, max([box[5] + box[2] for box in data[73]]))
    for box in data[73][:frame+1]:  # frame 번째까지의 박스만 그립니다.
        color_scale = box[7] / 10
        ax.bar3d(box[3], box[4], box[5], box[0], box[1], box[2], alpha=0.8,
                 color=plt.cm.Blues(color_scale))
    plt.title('Best Ratio (weight : 5) - Total 3D Box Plot')
ani = FuncAnimation(fig, update, frames=len(data[73]), interval=1200, repeat=False)
plt.show()
```

#### 밀도가 높고(0.6 초과) 배송지 번호(8 미만) 낮은 것 -> 팔레트 중간 
#### 밀도가 높고(0.6 초과) 배송지 번호 높은 것(8 초과) -> 팔레트 맨 아래
##### 밀도가 낮고(0.6 미만) 배송지 번호 낮은 것(8 미만) -> 팔레트 맨 위
##### 밀도가 높고(0.6 미만) 배송지 번호 높은 것(8 초과) -> 팔레트 중간

```python
#### 밀도가 높고(0.6 초과) 배송지 번호(7 이하) 낮은 것 : 연한 빨간색
#### 밀도가 높고(0.6 초과) 배송지 번호 높은 것(7 초과)  :빨간 색
##### 밀도가 낮고(0.6 이하) 배송지 번호 낮은 것(7 이하) : 파란색
##### 밀도가 높고(0.6 이하) 배송지 번호 높은 것(7 초과) : 연한 파란색

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import Normalize

# 3D 그래프 초기화
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 박스 정보를 이용하여 3D 상자 그리기
for box in data[84]:  # 첫 번째 에피소드 정보만 사용
    # 밀도 및 배송지 번호에 따른 색상 결정
    density = box[8]
    delivery_number = box[7]
    if density > 0.6 and delivery_number > 7:
        color = 'red'  # 밀도가 높고 배송지 번호 높은 것
    elif density > 0.6 and delivery_number <= 7:
        color = 'lightcoral'  # 밀도가 높고 배송지 번호 낮은 것
    elif density <= 0.6 and delivery_number > 7:
        color = 'lightblue'  # 밀도가 낮고 배송지 번호 높은 것
    else:
        color = 'blue'  # 밀도가 낮고 배송지 번호 낮은 것

    ax.bar3d(box[3], box[4], box[5], box[0], box[1], box[2], alpha=0.8, color=color)

# 축 레이블과 타이틀 설정
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Box Plot with Density and Delivery Number Color Coding')

plt.show()
```
