### #밀도, 배송지 함께 고려한 3d 시각화 (팔레트 맨 위 ~ 맨 아래까지 연한색 ~ 진한색으로 되는 방향이어야 함.)

#### 밀도가 높고(0.6 초과) 배송지 번호(8 미만) 낮은 것 -> 팔레트 중간 
#### 밀도가 높고(0.6 초과) 배송지 번호 높은 것(8 초과) -> 팔레트 맨 아래
##### 밀도가 낮고(0.6 미만) 배송지 번호 낮은 것(8 미만) -> 팔레트 맨 위
##### 밀도가 높고(0.6 미만) 배송지 번호 높은 것(8 초과) -> 팔레트 중간

```python
#박스 별 밀도의 분포 확인 : 밀도가 높고 낮음을 선정하기 위해, 즉 threshold 찾기 위해
# for sublist in data[0]: #에피소드 0번의 밀도 data만 확인
#     print(sublist[-1])
density_values = [sublist[-1] for sublist in data[84]]
density_values
```

```python
plt.figure(figsize=(10, 5))
plt.hist(density_values, bins=10, color='skyblue', edgecolor='black')
plt.title("Episode 84 data's density distribution")
plt.xlabel("Density")
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#동적으로 박스의 개수, 밀도는 계속 바뀌므로 걍 위 그래프 보고 len(density_values)/2 확인 후, 
#중간에 위치한 애의 밀도를 threshold로 선정해라.
```

```python
len(density_values)/2 #18번째 박스의 밀도 : 0.6으로 선정
```

```python
#여기선 *밀도* threshold=0.6으로 선정함.
```

```python
#박스 별 배송지 번호의 분포 확인 : 배송지 번호가 높고 낮음을 선정하기 위해, 즉 threshold 찾기 위해
# for sublist in data[0]: #에피소드 0번의 배송지 번호 data만 확인
#     print(sublist[-1])
bae_values = [sublist[-2] for sublist in data[84]]
bae_values
```

```python
plt.figure(figsize=(10, 5))
plt.hist(bae_values, bins=10, color='skyblue', edgecolor='black')
plt.title("Episode 84 data's bae_values distribution")
plt.xlabel("bae_values")
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
len(bae_values)/2 #18번째 박스의 bae_values : 7로 선정
```

```python
#### 밀도가 높고(0.6 초과) 배송지 번호(7 이하) 낮은 것 : 연한 빨간색
#### 밀도가 높고(0.6 초과) 배송지 번호 높은 것(7 초과)  :빨간 색
##### 밀도가 낮고(0.6 이하) 배송지 번호 낮은 것(7 이하) : 파란색
##### 밀도가 높고(0.6 이하) 배송지 번호 높은 것(7 초과) : 연한 파란색

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import Normalize

# 3D 그래프 초기화
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 박스 정보를 이용하여 3D 상자 그리기
for box in data[84]:  # 첫 번째 에피소드 정보만 사용
    # 밀도 및 배송지 번호에 따른 색상 결정
    density = box[8]
    delivery_number = box[7]
    if density > 0.6 and delivery_number > 7:
        color = 'red'  # 밀도가 높고 배송지 번호 높은 것
    elif density > 0.6 and delivery_number <= 7:
        color = 'lightcoral'  # 밀도가 높고 배송지 번호 낮은 것
    elif density <= 0.6 and delivery_number > 7:
        color = 'lightblue'  # 밀도가 낮고 배송지 번호 높은 것
    else:
        color = 'blue'  # 밀도가 낮고 배송지 번호 낮은 것

    ax.bar3d(box[3], box[4], box[5], box[0], box[1], box[2], alpha=0.8, color=color)

# 축 레이블과 타이틀 설정
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Box Plot with Density and Delivery Number Color Coding')

plt.show()
```
