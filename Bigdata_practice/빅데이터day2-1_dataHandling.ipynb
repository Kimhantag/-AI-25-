#### Q2. 아래는 운동에 따른 혈당 산소 요구량 데이터이다. pivot을 이용하여 GENDER와 AGEGROUP별 WEIGHT의 평균을 아래와 같이 변환하라.
- 데이터 : FITNESS.CSV(변수:10개, 자료 수 : 31개)

```python
fit = pd.read_csv("./실습파일/BigData분석기초/FITNESS.csv", encoding = "euc-kr")
fit.head()
```

```python
df1 = fit.groupby(['GENDER', 'AGEGROUP'], as_index=False).agg({'WEIGHT':'mean'})
df1
```

```python
df1.pivot(index='GENDER', columns='AGEGROUP', values='WEIGHT').round(3)
```

```python
df1.pivot(index='AGEGROUP',columns='GENDER', values='WEIGHT').round(3)
```
